1\. **PCA on 3D dataset**

* Generate a dataset with 3 features each with N entries (N being ${\cal O}(1000)$). With $N(\mu,\sigma)$ the normali distribution with mean $\mu$ and $\sigma$  standard deviation, generate the 3 variables $x_{1,2,3}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues of the covariance matrix of the dataset
* Find the eigenvectors and eigenvalues using SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained.
* Redefine the data in the basis yielded by the PCA procedure
* Plot the data points in the original and the new coordiantes as a set of scatter plots. Your final figure should have 2 rows of 3 plots each, where the columns show the (0,1), (0,2) and (1,2) proejctions.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg as la

def function(N):
    mean1, std1, std2 = 0, 1, 3
    x1 = np.random.normal(loc = mean1, scale = std1, size = N)
    x2 = x1 + np.random.normal(loc = mean1, scale = std2, size = N)
    x3 = 2 * x1 + x2
    return x1, x2, x3

N = 1000
x1, x2, x3 = function(N)
X = np.array([x1, x2, x3])
X_saved = np.copy(X)
covariance_matrix = np.cov(X)
#print('The covariance matrix is: \n', covariance_matrix)

l, V = la.eig(covariance_matrix)
l, V
#print('*', X)
#X[[0,1]] #prime due colonne dell'array
```

```python
U, d, Vt = la.svd(covariance_matrix)

D = np.zeros((U.shape))
for i in range(len(D)):
    D[i, i] = d[i]
cov = np.dot(U, np.dot(D, Vt))
print('Covariance matrix with SVD: \n ', cov)
print('Eigenvalues: \n', Vt.transpose())
print('Eigenvectors: \n', U)
```

```python
#Ora plottiamo i dati con autovettori ed autovalori
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, x3)
scale_factor = 1
#for li, vi in zip(l, V):
    #print(vi)
 #   ax.plot([0, scale_factor * li*vi[0]], [0, scale_factor * li*vi[1]], [0, scale_factor * li*vi[2]], c = 'red')

#Considero soltanto due variabili e cerco di scomporle 

plt.figure(figsize=(10,10))

plt.subplot(2, 2, 1)
plt.scatter(X[0,:], X[1,:])
plt.title('Variables x1 and x2')
for li, vi in zip(l, V.T):
    plt.plot([0, scale_factor*li*vi[0]], [0, scale_factor*li*vi[1]], 'r-', lw = 2)

plt.subplot(2, 2, 2)
plt.scatter(X[1,:], X[2,:])
plt.title('Variables x2 and x3')
for li, vi in zip(([l[0], l[2]]), ([V.T[0], V.T[2]])):
    plt.plot([0, scale_factor*li*vi[1]], [0, scale_factor*li*vi[2]], 'r-', lw = 2)
             
plt.subplot(2, 2, 3)
plt.scatter(X[0,:], X[2,:])
plt.title('variables x1 and x3')
for li, vi in zip(l[1::], V.T[1::]):
    plt.plot([0, scale_factor*li*vi[0]], [0, scale_factor*li*vi[2]], 'r-', lw = 2)
plt.show()
```

Osserviamo che nella rappresentazione di x1 ed x3 ed in x2 ed x3, l'autovettore non segue la reale rappresentazione dei dati. Questo significa che, essendo x3 linearmente dipendente rispetto ad x1 ed x2, allora  possibile eliminare questa dimensione.

```python
#What percent of the total dataset's variability is explained
#by the principal components? Given how the dataset was constructed, 
#do these make sense? Reduce the dimensionality of the system so that 
#at least 99% of the total variability is retained.

#vediamo quali autovalori rappresentano il 99 % del dataset 

somma = np.sum(l)
frac_x1, frac_x2, frac_x3 = l[0]/somma * 100 , l[1]/somma * 100 , l[2]/somma * 100

print(frac_x1, '%', frac_x2, '%', frac_x3, '%')

#Posso eliminare la terza dimensione x3


i = l.argsort()[::-1] #trucchetto per riordinare la matrice in ordine decrescente, di modo che gli autovettori siano riferiti 
                      #al dato specifico 
V = V[:, i]
#Ridefinisco nella base del PCA
#V = V[:, 1]  #prima e terza colonna 
#V = V[[0,2], :] #prima e terza riga 

Xp = np.dot(V[:,:2].T, X) #shape 


Xp = np.vstack((Xp, np.zeros(Xp.shape[1])))

#for li, vi in zip(l, np.diag([1]*2)):
 #   plt.plot([0, scale_factor*li*vi[0]], [0, scale_factor*li*vi[1]], 'r-', lw=2)

plt.figure(figsize=(22,22))
plt.subplot(331)
plt.scatter(Xp[0,:], Xp[1,:], alpha=0.2)
plt.title('X1 and X2 after dimension reduction')

plt.subplot(332)
plt.scatter(Xp[0,:], Xp[2,:], alpha = 0.2)
plt.title('X1 and X3 after dimension reduction')

plt.subplot(333)

plt.scatter(Xp[1,:], Xp[2,:], alpha = 0.2)
plt.title('X2 and X3 after dimension reduction')

plt.subplot(334)
plt.scatter(X[0,:], X[1,:], alpha = 0.2)
plt.title('X1 and X2')

plt.subplot(335)
plt.scatter(X[0, :], X[2,:], alpha = 0.2)
plt.title('X1 and X3')

plt.subplot(336)
plt.scatter(X[1, :], X[2, :], alpha = 0.2)
plt.title('X2 and X3')
```

Repeat the PCA procedure and compare the results with what you obtained before

```python
def add_noise():
    std1 = 1
    std2 = 3
    std = np.linspace(std1, std2, 10)
    mean = 0
    N = 1000
    std_matrix = [sigma /50 for sigma in std]
    
    x_array = [np.random.normal(loc = mean, scale = std, size = N) for std in std_matrix]
    
    return x_array

x = add_noise()

X2 = np.concatenate((X_saved, x))
#genero la distribuzione
```

```python
#applico la pca
cov_matrix = np.cov(X2)

l, V = la.eig(cov_matrix)

i = l.argsort()[::-1] #trucchetto per riordinare la matrice in ordine decrescente, di modo che gli autovettori siano riferiti 
                      #al dato specifico 
V = V[:, i]

somma = np.sum(l)

frac = [li / somma * 100 for li in l]
print(frac) #osservo la scomposizione in percentuale degli autovalori che ho ottenuto ora 
```

3 \. **Looking at an oscillating spring** (optional)

Imagine you have $n$ cameras looking at a spring oscillating along the $x$ axis. Each  camera record the motion of the spring looking at it along a given direction defined by the pair $(\theta_i, \phi_i)$, the angles in spherical coordinates. 

Start from the simulation of the records (say ${\cal O}(1000)$) of the spring's motion along the x axis, assuming a little random noise affects the measurements along the $y$. Rotate such dataset to emulate the records of each camera.

Perform a Principal Component Analysis on the thus obtained dataset, aiming at finding the only one coordinate that really matters.


```python
N = 1000
a = 0
b = 10 

#definisco le matrici di rotazione
def oscillating(t):
    A = 1 #oscillation amplitude
    omega = 2 * np.pi
    phi = 0
    return A * np.cos(omega * t + phi)

theta = np.linspace(0, np.pi / 2, N)
phi = np.linspace(0, np.pi, N)

t = np.linspace(0, 100, N)
x = [oscillating(t) for t in t] #= x posizione nel tempo della molla 

y = [np.random.normal(loc = 0, scale = max(x) / 100) for _ in range(N)]

z = np.zeros(len(x))

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d', alpha = 0.5)

ax.scatter(x, y, z )
ax.quiver(0,0, 0, 1, 0, 0, color='red', label='X')
ax.quiver(0,0, 0, 0, 1, 0, color='green', label='Y')
ax.quiver(0,0, 0, 0, 0, 1, color='blue', label='Z')

#ax.w_xaxis.line.set_lw(0.)
#ax.w_yaxis.line.set_lw(0.)
#ax.w_zaxis.line.set_lw(0.)

ax.view_init(elev=40, azim=30)
ax.set_title('Motion of the spring')

X = np.array([x, y, z]) #ottengo l'array completo 
```

```python
#Performo la PCA sul dataset che ho ottenuto 

covariance_matrix = np.cov(X)
l, V = la.eig(covariance_matrix)


scale_factor = 5

plt.figure(figsize=(15,20))
plt.subplot(331)
plt.scatter(X[0], X[1])
plt.axhline(0, color='black', linewidth=0.8)  # Linea per l'asse x
plt.axvline(0, color='black', linewidth=0.8)  # Linea per l'asse y

plt.xlabel('X axis')
plt.ylabel('Y axis')


i = l.argsort()[::-1]  

V = V[:, i]


for li, vi in zip(l, V.T):
    plt.plot([0, scale_factor*li*vi[0]], [0, scale_factor*li*vi[1]], 'r-', lw = 2)
    
plt.title('Projection on the xy plane')

plt.subplot(332)
plt.scatter(X[0], X[2])
for li, vi in zip(([l[0], l[2]]), ([V.T[0], V.T[2]])):
    plt.plot([0, scale_factor*li*vi[0]], [0, scale_factor*li*vi[2]], 'r-', lw = 2)
    
plt.xlabel('X axis')
plt.ylabel('Z axis')
plt.title('Projection on the xz plane')

plt.subplot(333)
plt.scatter(X[1], X[2])
for li, vi in zip(l[1::], V.T[1::]):
    plt.plot([0, scale_factor*li*vi[1]], [0, scale_factor*li*vi[2]], 'r-', lw = 2)
    
plt.xlabel('Y axis')
plt.ylabel('Z axis')
plt.title('Projection on the yz plane')
```

```python
import requests
#non so perrch ma non mi scarica il file, lo scarico a mano 

def scarica_file_da_url(url, percorso_destinazione):
    try:
        # Effettua una richiesta GET al server per ottenere il contenuto del file
        risposta = requests.get(url)
        
        # Verifica se la richiesta  andata a buon fine (status code 200)
        if risposta.status_code == 200:
            # Apre il file in modalit scrittura binaria e scrive il contenuto ottenuto
            with open(percorso_destinazione, 'wb') as file:
                file.write(risposta.content)
            print("Il file  stato scaricato con successo.")
        else:
            print(f"Errore nella richiesta HTTP. Status code: {risposta.status_code}")
    except Exception as e:
        print(f"Si  verificato un errore: {str(e)}")

# Specifica l'URL del file che desideri scaricare
url_file = 'https://archive.ics.uci.edu/ml/machine-learning-databases/magic/magic04.data'

# Specifica il percorso locale in cui desideri salvare il file
percorso_locale = 'C:\\Users\\adadi\\OneDrive\\Desktop\\exe_lab_com\\LaboratoryOfComputationalPhysics_Y6-main'

# Chiama la funzione per scaricare il file
scarica_file_da_url(url_file, percorso_locale)
```
