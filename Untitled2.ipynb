```python
#credit card exercise
data = 'credit_card.dat' #devo caricare questi dati 


with open(data, 'rb') as file:
    card = []
    for line in file:
        string = line[:-2]
        byte = []
        for i in range(0, len(string) - 6, 6):
            block = chr(int(string[i:i+6], 2))
            byte.append(block)
        card.append(byte)

line = ''
for i in range(0, len(card)):
    for j in range(len(card[i])):
        line += str(card[i][j])
        if j == len(card[i])- 1:
            line += '\n'
print(line)
```

```python
#Let's plot the number of counts per TDC channel 
count0 = 0
count1 = 0
fpga1 = []
fpga2 = []

s = 0
for i in df['FPGA']:
    
    if i == 0:
        count0 += 1
        fpga1.append(df['TDC_CHANNEL'].iloc[s])
    else: 
        count1 += 1
        fpga2.append(df['TDC_CHANNEL'].iloc[s])
    s += 1
#plt.scatter(count0, fpga1)

channel1, count1_ = np.unique(fpga1, return_counts = True)
channel2, count2_ = np.unique(fpga2, return_counts = True)

plt.subplot(211)
plt.scatter(channel1, count1_)
plt.title('Counts per channel for fpga 1')

plt.subplot(212)
plt.scatter(channel2, count2_)
plt.title('Counts per channel for fpga 2')
```

```python
#Visualization 

N = 100
mean = 0
std = 1
x = np.random.normal(loc = mean, scale = std, size = N)

nbin =np.sqrt(N)

y, bordi_bin, _ = plt.hist(x, bins =int(nbin))
plt.xlabel('X distribution')
plt.ylabel('Frequencies')
```

```python
plt.hist(x, bins =int(nbin))
mean_bin = [( bordi_bin[i+1] + bordi_bin[i] )/ 2 for i in range(len(bordi_bin) - 1)]

error =np.sqrt(y) #errpre secondo la definizione di poisson 
plt.errorbar(mean_bin, y, yerr = error, fmt='o',color = 'red')
```

```python
from scipy.stats import norm


std_x = 1.06 * x.std() * x.size ** (1/5)

n_points = 200
x_axis = np.linspace(x.min() - 4 * std_x, x.max() + 4 * std_x, n_points)
#xi = [np.random.normal(loc = i, scale = std_x, size = n_points) for i in x]

pdf  = [norm.pdf(x_axis, loc = i, scale = std_x) for i in x ]

for i in pdf:
    plt.plot(x_axis, i)
    
```

```python
#Calcoliamo l'integrale delle gaussiane 
from scipy.integrate import trapz 

sum_gaussian = np.sum(pdf, axis = 0)
area_under_curve = trapz(sum_gaussian, x_axis)
histogram_integral = trapz(y, mean_bin)

normalized_sum_gaussian =  sum_gaussian  / area_under_curve  * histogram_integral

print('Histogram integral: ', histogram_integral)
print('Area under curve: ', area_under_curve)
print('Normalized integral: ', np.sum(normalized_sum_gaussian))
```
