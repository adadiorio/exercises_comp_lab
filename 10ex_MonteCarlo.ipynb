${\rm Tl}^{208}$ decays to ${\rm Pb}^{208}$ with a half-lieve of 3.052 minutes. Suppose to start with a sample of 1000 Thallium atoms and 0 of Lead atoms.

* Take steps in time of 1 second and at each time-step decide whether each Tl atom has decayed or not, accordingly to the probability $p(t)=1-2^{-t/\tau}$. Make a plot showing the number of atoms that have not decayed as a function of time

```python
import numpy as np
import matplotlib.pyplot as plt 

tau = 3.052 * 60 #in seconds  

def probability(t):
    return 1 - 2 ** (-t/tau) #probabilità che un atomo sia decaduto

N_ti = 1000
N_Le = 0
time, TIME, TI, LE = 0, [0], [N_ti], [N_Le]
counter = 0

while N_ti > 0:
    Ti_atoms = [np.random.uniform(0, 1) for i in range(N_ti)] #associo ad ogni atomo un valore 
                                                             #di probabilità, che rappresenta la 
                                                             #probabilità che l'atomo sia decaduto oppure no
    for i in range(len(Ti_atoms)): #ad ogni step vedo quanti atomi sono decaduti
        if Ti_atoms[i] < probability(1):
            N_ti -= 1 #l'atomo decade
            N_Le += 1 #incremento di uno il numero di atomo di Pb
           
        else:
            None
            
        
    time += 1 #ad ogni istante valuto se tutti gli atomi sono decaduti oppure no 
        
    TI.append(N_ti)
    LE.append(N_Le)
    TIME.append(time)

plt.scatter(TIME, TI, label = 'Ti atoms', s = 0.5)
plt.scatter(TIME, LE, label = 'Le atoms', s = 0.5)
plt.axvline(x = tau, linestyle = '--', label = 'halflife', color = 'r')
plt.xlabel('Time [s]')
plt.ylabel('Number of atoms')
plt.legend()
plt.title('Evolution over time')
```

2\. **Rutherford Scattering**

The scattering angle $\theta$ of $\alpha$ particles hitting a positively charged nucleus of a Gold atom ($Z=79$) follows the rule:

$$
\tan{\frac{1}{2} \theta} = \frac{Z e^2} {2\pi \epsilon_0 E b}
$$

where $E=7.7$ MeV and $b$ beam is the impact parameter. The beam is represented by a 2D gaussian distribution with $\sigma=a_0/100$ for both coordinates ($a_0$ being the Bohr radius). Assume 1 million $\alpha$ particles are shot on the gold atom.

Computing the fraction of particles that "bounce back",i.e. those particle whose scattering angle is greater than $\pi/2$ (which set a condition on the impact parameter $b$)

```python
a0 = 5 * 10 ** -11 #m
sigma = a0/100
E = 7.7 * 10 **6 * 1.602 * 10 **-19
Z = 79
N =  10**6 #numero di particelle inviate 

def get_beam(sigma, N): #N numero di particelle sparate, numero di punti che compongono il fascio 
    mean = 0 #valore atteso 0, centro la targhetta 
    x = np.random.normal(loc = mean, scale = sigma, size = N)
    y = np.random.normal(loc = mean, scale = sigma, size = N)
    return x, y

#Ogni coppia (x,y) è una particella nello spazio, a cui corrisponde un dato angolo di scattering 
def tan_scattering_angle(x, y):
    #b = np.sqrt(x**2 + y**2)
    epsilon0 = 8.854 * 10 **-12 #C^2 / N * m^2
    e = 1.602 * 10**-19  #MeV
    b = Z * e**2 / ( 2 * np.pi * epsilon0 * E)
    
    return b

#Genero N particelle e ciascuna sarà caratterizzata da una distribuzione x,y
X, Y = get_beam(sigma, N) #ogni particella è individuata da una coordinata (x, y) distribuita secondo una gaussiana

hits = 0
for i in range(N):
    xi, yi = X[i], Y[i]
    b = np.sqrt(xi**2 + yi**2)
    if b  < tan_scattering_angle(xi, yi): #rimbalzano 
        hits += 1

fraction = hits / N
print('Fraction of bouncing back particles: ', fraction * 100, '%')
```

3\. **Monte Carlo integration: hit/miss vs mean value method**

Consider the function 

$$f(x) =\sin^2{\frac{1}{x(2-x)}}$$

* Compute the integral of $f(x)$ between 0 and 2 with the hit/miss method. Evaluate the error of your estimate
* Repeat the integral with the mean value method. Evaluate the error and compare it with the previous one

```python
def function(x):
    return np.sin(1/(x*(2-x)))**2

N = 10000
hits = 0

X, Y = [], []
for i in range(N):
    x = np.random.random() * 2
    y = np.random.random() 
    
    X.append(x)
    Y.append(y)
    
    if y < function(x):
        hits += 1
        
integral = hits / N * 2

plt.scatter(X, Y, s = 0.3)
std_dev = np.sqrt(np.std(X) + np.std(Y))
func = [function(x) for x in X]
plt.scatter(X, func, s = 0.5)

print('Integral with monte carlo method: ', integral, '+-', std_dev/np.sqrt(N))
```

4\. **Monte Carlo integration in high dimension**

* Start of by computing the area of a circle of unit radius, by integrating the function 

$$
f(x,y)=
\left\{
\begin{array}{ll}
      1 & x^2+y^2\le 1 \\
      0 & {\rm elsewhere}
\end{array} 
\right.
$$

* Generalize the result for a 10D sphere



```python
def function(x,y):
    if x**2 + y**2 <= 1:
        return 1
    else:
        return 0

N = 10000
X = []
Y = []
hits = 0
for i in range(N):
    x = np.random.uniform(-1, 1) 
    y = np.random.uniform(-1, 1) 
    
    X.append(x)
    Y.append(y)
    #print(function(x, y))
    if x**2 + y**2 <= function(x, y):
        hits += 1
        
Integral = 4 * (hits / N)
print(Integral, '+-', np.sqrt(np.std(X)**2 + np.std(Y)**2)/N)

circ = [function(i, j) for i, j in zip(X, Y)]
plt.scatter(X, Y)
plt.scatter(X, circ)
```
